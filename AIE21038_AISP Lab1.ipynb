```python
import librosa
import librosa.core
import librosa.display
import numpy as np
import IPython.display as ipd
import matplotlib.pyplot as plt
y,sr=librosa.load('ais.wav')
librosa.display.waveshow(y)
```

```python
print("length of signal is:",len(y))
print("Magnitude is:",np.min(y),"to",np.max(y))
```

```python
start_time=5
end_time=12
segment=y[int(start_time*sr):int(end_time*sr)]
ipd.Audio(segment, rate=sr)
```

```python
segment_1_start = int(1 * sr)
segment_1_end = int(4 * sr)
segment_2_start = int(5 * sr)
segment_2_end = int(9 * sr)
segment_1 = y[segment_1_start:segment_1_end]
segment_2 = y[segment_2_start:segment_2_end]
print("Playing Segment 1:")

plt.figure(figsize=(10, 5))
librosa.display.waveshow(segment_1)
plt.title('Segment-1')
plt.show()
ipd.Audio(segment_1, rate=sr)
```

```python
print("Playing Segment 2:")

plt.figure(figsize=(10, 5))
librosa.display.waveshow(segment_2)
plt.title('Segment-2')
plt.show()
ipd.Audio(segment_2, rate=sr)
```
